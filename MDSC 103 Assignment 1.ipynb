```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.optimize import minimize
from mpl_toolkits.mplot3d import Axes3D
```

# Surface Plot

```python
f = lambda x1, x2: x1**3 +x2**3 - 3*x1*x2
x1 = np.linspace(-5,5,100)
x2 = np.linspace(-5,5,100)
#x2 = x1.copy()
x1, x2 = np.meshgrid(x1,x2)
F = f(x1,x2)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(1,2,1,projection='3d') # set up the axis as a 3D plot 
ax.scatter3D(1,1,f(1,1),c="black",marker="*",s=100)
surf = ax.plot_surface(x1,x2,F, cmap= 'gist_rainbow')
fig.colorbar(surf,shrink=0.4, aspect=10)
# set axes label
ax.set_xlabel('x1', labelpad=10)
ax.set_ylabel('x2', labelpad=30)
ax.set_zlabel('f(X)', labelpad=10)

plt.show()
```

```python
# minimize function without constraints

f = lambda x: x[0]**3 +x[1]**3 - 3*x[0]*x[1]
x0 = [2,1]
res = minimize(f,x0)
res
```

```python
# minimize function with constraints

f = lambda x: x[0]**3 +x[1]**3 - 3*x[0]*x[1]
cons = ({'type': 'ineq','fun': lambda x: 3*x[0]+x[1]-3}, # All constraints are greater than  3x1 + x2 <3
        {'type': 'ineq','fun': lambda x: 5*x[0]-3*x[1]-5}, #  All constraints are greater than 5x1 - 3x2 < 5
        
        {'type': 'eq','fun': lambda x: 3*x[0]+x[1]-3}, # All constraints are equal 3x1 + x2 = 3
        {'type': 'eq','fun': lambda x: 5*x[0]-3*x[1]-5}#All constraints are equal 5x1 - 3x2 < 5
        
        )

bnds = ((0,None),(0,None)) 
x0 = (2,0)
res = minimize(f,x0, bounds=bnds, constraints= cons)
res2 = minimize(f,x0)
res2
```

```python
# Plotting Function and constraints on a 2d plane

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D


f = lambda x1,x2: x1**3 + x2**3 - 2*x1*x2   #Function
f1 = lambda x1,x2: -3*x1-x2+3         # Constraint1
f2 = lambda x1,x2: -5*x1+3*x2+5       # Constraint2

x1 = np.linspace(0,5,100)
x2 = np.linspace(0,5,100)

X, Y = np.meshgrid(x1,x2)
F = f(X,Y)
F1 = f1(X,Y)
F2 = f2(X,Y)


plt.plot(X,Y,F, color = 'r')
plt.plot(X,Y,F1, color = 'b')
plt.plot(X,Y,F2, color = 'g')

plt.show()
```
