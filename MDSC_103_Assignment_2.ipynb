## **Plotting the graphs of the y=x^2 and y=x^3 and convergence functions using Gradient Descent Method**

```python
#plotting the graph for the y = x^2 function
import numpy as np
import matplotlib.pyplot as plt

f = lambda x:x**2
z = np.linspace(-5,5,100)
plt.plot(z,f(z))
plt.title('Plot of y = x^2 function')
plt.xlabel('x values from -5 to +5')
plt.ylabel('x^2 values')
plt.show()
```

```python
# Define the function
f = lambda x: x**2

# Generate data for x and y values
z = np.linspace(-5, 5, 100)
y = f(z)

# Create the figure and axis
fig, ax = plt.subplots()
ax.plot(z, y, label='$f(x) = x^2$')

# Gradient Descent parameters
x0 = 3
learning_rate = 0.1
tolerance = 1e-5

# Gradient Descent function
GD = lambda x0: x0 - learning_rate * 2 * x0

# Perform Gradient Descent iterations
for i in range(100):
    x1 = GD(x0)
    if abs(x0 - x1) < tolerance:
        print("Convergence at:", x1)
        break
    ax.scatter(x1, f(x1), color='r')

    x0 = x1

# Add labels and title to the plot
ax.set_title('Convergence of $f(x) = x^2$ using Gradient Descent')
ax.set_xlabel('x values')
ax.set_ylabel('y values')

# Highlight the final point of convergence
ax.scatter(x1, f(x1), color='g', marker='o', label='Converged Point')
ax.annotate(f'({x1:.2f}, {f(x1):.2f})', (x1, f(x1)), textcoords="offset points", xytext=(0,10), ha='center')

# Add legend
ax.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
def GD_fun(fun, df, ip, lr, tol):
    """
    Perform Gradient Descent optimization on a given function.

    Parameters:
    fun: The function to optimize.
    df: The derivative of the function.
    ip: The initial point for optimization.
    lr: The learning rate for optimization.
    tol: The convergence tolerance.
    """

    # Generate x values from -5 to 5
    x = np.linspace(-5, 5, 100)

    # Plot the original function f(x)
    plt.plot(x, fun(x))

    # Add gridlines to the plot
    plt.grid()

    # Label for x-axis
    plt.xlabel('-5 to +5 values')

    # Label for y-axis
    plt.ylabel('Squares of x values')

    # Define the Gradient Descent update formula using a lambda function
    GD = lambda x0, lr: x0 - (lr) * df(x0)

    # Loop for performing Gradient Descent iterations
    for i in range(100):
        # Update x1 using the Gradient Descent formula
        x1 = GD(ip, lr)

        # Plot the current iteration point
        plt.scatter(x1, fun(x1), color='r')

        # Set the title for the plot
        plt.title('Convergence of function using Gradient Descent method')

        # Check if convergence criterion is met
        if np.abs(ip - x1) < tol:
            print("Convergence at:", x1)
            break

        # Update the initial point for the next iteration
        ip = x1

    # Display the plot
    plt.show()
```

```python
# Define the function f(x) = x^2 using a lambda function
fun = lambda x: x**2

# Define the derivative of the function, which is df(x) = 2x
df = lambda x: 2*x

# Call the GD_fun function with specified parameters
# The function is being optimized: f(x) = x^2
# Its derivative: df(x) = 2x
# Initial point: ip = 4
# Learning rate: lr = 0.1
# Convergence tolerance: tol = 1e-5
GD_fun(fun, df, 4, 0.1, 1e-5)
```

## **y=x^3**

```python
# Define the function f(x) = x^3 using a lambda function
fun = lambda x: x**3

# Define the derivative of the function, which is df(x) = 3x^2
df = lambda x: 3*x**2

# Call the GD_fun function with specified parameters
# The function is being optimized: f(x) = x^3
# Its derivative: df(x) = 3x^2
# Initial point: ip = 2
# Learning rate: lr = 0.1
# Convergence tolerance: tol = 1e-5
GD_fun(fun, df, 2, 0.1, 1e-5)
```
